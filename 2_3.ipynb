## 2.3.1 Joint Probability Distributions
Joint probability is the probability of two or more events happening together.
A joint probability distribution shows a probability distribution for two or more random variables.

The pmf of a single discrete random variable $X$ can be extended to two variables $X,Y$ to describe how much probability mass is placed on each possible pair of values. Specifically:
\begin{equation}
p(x,y)=P(X=x\text{ and }Y=y)
\end{equation}
where $p(x,y)\geq0$ and the sum of all pmf values is equal to 1.

Joint continuous distributions can be calculated similarly.
### Independent Random Variables
Two random variables $X$ and $Y$ are independent if for all pair of $x$ and $y$ values $p(x,y)=p_x(x)\cdot p_y(y)$ (and likewise for continuous random variables).

## 2.3.3 Random Samples
A random sample is a randomly selected subset of a population.

Random variables $X_1,X_2,...,X_n$ are said to form a simple random sample of size $n$ if
1. The random variables are independent
2. each random variable have the same probability distribution

In general, we can draw conclusions about the population mean μ using the sample mean $\bar{X}=(X_1+...+X_n)/n$.
### Central Limit Theorem
The Central Limit Theorem (CLT) says that a normalized sum of independent random variables will tend towards a normal distribution even if the variables are not normally distributed.
This is an important theorem because it implies that methods that work for normal distributions can be applied to problems that involve other distributions.

We can demonstrate this by generating a number of random numbers in the range -100, 100, calculating the mean

```python
import numpy
import matplotlib.pyplot as plt
 
# number of sample
num = [1, 10, 50, 100] 
# list of sample means
means = [] 
 
# Generating 1, 10, 30, 100 random numbers from -100 to 100
# taking their mean and appending it to list means.
for j in num:
    # Generating seed so that we can get same result every loop
    numpy.random.seed(1)

    # generate j random numbers in [-100, 100]
    # calculate the sample mean of those j numbers
    # repeat the above 1000 times
    # store results in x and append to means
    x = [numpy.mean(numpy.random.randint(-100, 100, j)) for i in range(1000)]
    means.append(x)

k = 0
# plotting all the means in one figure
fig, ax = plt.subplots(2, 2, figsize =(8, 8))
for i in range(0, 2):
    for j in range(0, 2):
        # Histogram for each x stored in means
        ax[i, j].hist(means[k], 20, density = True)
        ax[i, j].set_title(label = "sample size=" + str(num[k]))
        k += 1
plt.show()
```
