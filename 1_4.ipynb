## 1.4.1 Singular Value Decomposition
The Singular Value Decomposition of a matrix has many applications, including data science and image processing.
It takes a `m x n` matrix `A` and decomposes it into three matrices: `U`, an `m x m` orthogonal matrix, Σ, a `m x n` matrix where the diagonal entries in `D` are the first `r` singular values of `A`, and an `n x n` orthogonal matrix `V` such that `A=U`Σ`V^T`.
The diagonal entries in Σ are the singular values for `A`.
numpy has functionality for computing SVD's:


```python
import numpy as np
A = np.random.normal(size=[20,18])
U, sigma, V = np.linalg.svd(A, full_matrices=False)
A_a = np.matmul(np.matmul(U, np.diag(sigma)), V)
print(np.std(A), np.std(A_a), np.std(A - A_a))
print("Is A close to A_a?", np.isclose(A, A_a).all())
```
